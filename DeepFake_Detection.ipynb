```python
import tensorflow as tf
import os

def preprocess_image(img, label):
    img = tf.cast(img, tf.float32) / 255.0  # Normalize the image to [0, 1]
    return img, label

dataset_dir_1 = '/content/deepfake_and_real_images/Dataset'
train_1 = tf.keras.preprocessing.image_dataset_from_directory(
    os.path.join(dataset_dir_1, 'Train'),
    image_size=(128, 128),
    batch_size=32,
    label_mode='int',
    color_mode="rgb",
    seed=42
)

val_1 = tf.keras.preprocessing.image_dataset_from_directory(
    os.path.join(dataset_dir_1, 'Validation'),
    image_size=(128, 128),
    batch_size=32,
    label_mode='int',
    color_mode="rgb",
    seed=42
)

test_1 = tf.keras.preprocessing.image_dataset_from_directory(
    os.path.join(dataset_dir_1, 'Test'),
    image_size=(128, 128),
    batch_size=32,
    label_mode='int',
    color_mode="rgb",
    seed=42
)

train_1 = train_1.map(preprocess_image)
val_1 = val_1.map(preprocess_image)
test_1 = test_1.map(preprocess_image)

for image, label in train_1.take(1):
    print(image.shape)
```

```python
import tensorflow as tf
import os
from sklearn.model_selection import train_test_split

dataset_dir_2 = '/content/faceforensics.zips/cropped_images'

image_paths_2 = []
for root, dirs, files in os.walk(dataset_dir_2):
    for file in files:
        if file.endswith(('.jpg', '.png', '.jpeg')):
            image_paths_2.append(os.path.join(root, file))

# Since all images in this dataset are fake, assign label 0 for fake images
labels_2 = [0] * len(image_paths_2)

# Split second dataset into two halves (50% for training, 50% for testing)
train_paths_2, test_paths_2 = train_test_split(image_paths_2, test_size=0.5, random_state=42)
train_labels_2 = [0] * len(train_paths_2)
test_labels_2 = [0] * len(test_paths_2)

train_2 = tf.data.Dataset.from_tensor_slices((train_paths_2, train_labels_2))
test_2 = tf.data.Dataset.from_tensor_slices((test_paths_2, test_labels_2))

def preprocess_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [128, 128])
    img = tf.cast(img, tf.float32) / 255.0
    return img, label

train_2 = train_2.map(preprocess_image).batch(32)
test_2 = test_2.map(preprocess_image).batch(32)

for image, label in train_2.take(1):
    print(image.shape)
```

```python
import tensorflow as tf
import os
from tensorflow.keras.preprocessing import image_dataset_from_directory

def preprocess_image(img, label):
    img = tf.cast(img, tf.float32) / 255.0  # Normalize the image to [0, 1]
    return img, label

dataset_dir_3 = '/content/dfdc-faces-of-the-train-sample'

train_3 = image_dataset_from_directory(
    os.path.join(dataset_dir_3, 'train'),
    image_size=(128, 128),
    batch_size=32,
    label_mode='int',
    color_mode="rgb",
    seed=42
)

val_3 = image_dataset_from_directory(
    os.path.join(dataset_dir_3, 'validation'),
    image_size=(128, 128),
    batch_size=32,
    label_mode='int',
    color_mode="rgb",
    seed=42
)

train_3 = train_3.map(preprocess_image)
val_3 = val_3.map(preprocess_image)

for image, label in train_3.take(1):
    print(f'Train 3 image shape: {image.shape}')

for image, label in val_3.take(1):
    print(f'Validation 3 image shape: {image.shape}')
```

```python
# Combine the datasets
train_data_combined = train_1.concatenate(val_1).concatenate(train_2).concatenate(train_3)
```

```python
for image, label in train_data_combined.take(1):
    print(f"Shape of combined dataset images: {image.shape}")
    print(f"Shape of labels: {label.shape}")
```

```python
model = models.Sequential()

model.add(layers.Conv2D(32,(3,3),activation="relu", padding='same',input_shape=(128,128,3)))
model.add(layers.Conv2D(32,(3,3),activation="relu", padding='same'))
model.add(layers.MaxPooling2D((2,2), strides=(2, 2)))


model.add(layers.Conv2D(64,(3,3),activation="relu", padding='same'))
model.add(layers.Conv2D(64,(3,3),activation="relu", padding='same'))
model.add(layers.MaxPooling2D((2,2), strides=(2, 2)))


model.add(layers.Conv2D(128,(3,3),activation="relu", padding='same'))
model.add(layers.Conv2D(128,(3,3),activation="relu", padding='same'))
model.add(layers.MaxPooling2D((2,2), strides=(2, 2)))

model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(128,activation="relu"))
model.add(layers.Dense(256,activation="relu"))
model.add(layers.Dense(1,activation="sigmoid"))

model.summary()
```

```python
model.compile('adam',loss="binary_crossentropy", metrics=['accuracy'])
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3,restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)

history = model.fit(train_data_combined,epochs=20,batch_size=32,callbacks=[early_stopping_cb,reduce_lr])
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.show()
```
